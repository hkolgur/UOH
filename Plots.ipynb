# Matplotlib Examples

```
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0, 5, 0.2)
plt.plot(t,2*t*t,label='2*t*t') # default is green continuous line
plt.plot(t,t**2,'ro',label='t**2') # red O dots
plt.plot(t,t**3,'b--',t,2*t,'g^',label=('2*t'))
#plt.plot(t,2*t,'g^',label='2*t')
plt.legend()    # only then the legends on each plot will display 
plt.xlabel("Default X values taken as Indices")
plt.ylabel("Taken as Y values")
plt.title("Test Plot")
plt.grid()  #plots a grid on backgorund
plt.show()

```

## Plt returns lines and control properties of each line

```
x1=[1,2,3,4]
y1=[8,6,5,4]
x2=[4,3,2,1]
y2=[8,6,5,4]
lines=plt.plot(x1,y1,x2,y2) #plt returns lines
plt.setp(lines[0],color='g',linewidth=4.0,linestyle=':')   # we can set properties for each of the line
plt.setp(lines[1],color='r',linewidth=2.0,linestyle='solid')
```

## Multiple Plots
### can go back to previous figure by specifiying the number of the figure you want to go and the particular subplot too
#matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True,FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)

```
plt.figure(1,figsize=(8,4),facecolor='lightgreen',edgecolor='b') #figsize(width,height)
plt.subplot(1,2,1) # says 1 row 2 columns and access first one
plt.grid()
plt.plot([1,2,3,4],[1,4,9,16],'b',linestyle='solid')
#ACCESS SECOND SUB PLOT
plt.subplot(1,2,2)# says 1 row 2 columns and access second plot 
plt.plot([11,12,13,14],[1,14,9,16],'r--')

#CREATE ONE MORE FIGURE 
plt.figure(2)
plt.plot([0,1,2,3,4],[0,.1,.0,.2,.1])

#CREATE THIRD FIGURE WITH 3 SUBPLOTS
plt.figure(3)
plt.subplot(3,1,1)
plt.plot([1,2,3,4],[1,4,9,16],'b',linestyle='solid')
plt.subplot(3,1,2)# says 1 row 2 columns and access second plot 
plt.plot([11,12,13,14],[1,14,9,16],'r--')
plt.subplot(3,1,3)# says 1 row 2 columns and access second plot 
plt.plot([11,12,13,14],[1,14,9,16],'go')

plt.figure(1)
plt.subplot(1,2,1)
plt.title("Traversed back 1st Sub")
plt.subplot(1,2,2)
plt.title("Traversed back 2nd Sub")

plt.figure(3)
plt.subplot(3,1,1)
plt.title("Returned Back to Third Figure")
```

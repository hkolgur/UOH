```python
#Given xt,h_prev,w1,w2,w3 ----> find ht 
```

```python
import numpy as np
def fn_gru_pred(xt,h_prev,w1,w2,w3):
  def fn_sig(x): return 1/(1+np.exp(-x))

  concat1=np.concatenate([h_prev,xt])

  r=fn_sig(w1@concat1)
  z=fn_sig(w2@concat1)

  concat2=np.concatenate([r*h_prev,xt])
  h_=np.tanh(w3@concat2)
  h=(1-z)*h_prev + z*h_
  return h.round(3)
```

```python
xt=np.array([-.26,-1.79])
h_prev=np.array([.05,.06])
w1=np.array([[.89,.1,.21,.71],
             [.06,.21,.21,.71]])
w2=np.array([[.1,.94,.21,.71],
             [.21,.42,.21,.71]])
w3=np.array([[.21,.71,.21,.71],
             [.06,.21,.21,.71]])


fn_gru_pred(xt,h_prev,w1,w2,w3)
```
